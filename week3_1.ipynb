<h1> reading table from wikipedia html page by pd.read_html method
    and displaying top 5 rows
 <h2> displaying 3 rows in dataframe

```python
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.head()
```

<h1>removing rows where Borough is not assigned 

```python
df=df[df['Borough']!='Not assigned']
df
```

<h2>joining neighborhoods with same postal code

```python
df.groupby(["Postal Code", "Borough"], as_index=False).agg(lambda x: ", ".join(x))
```

<h1>assigning borough as neighborhood where neighborhood is not assigng

```python
for index, row in df.iterrows():
    if row["Neighborhood"] == "Not assigned":
        row["Neighborhood"] = row["Borough"]
        
df.head()
```
